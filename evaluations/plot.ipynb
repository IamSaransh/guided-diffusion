```python

#grad_norm,param_norm,samples,step,train_acc@1,train_acc@1_q0,train_acc@1_q1,train_acc@1_q2,train_acc@1_q3,train_acc@5,train_acc@5_q0,train_acc@5_q1,train_acc@5_q2,train_acc@5_q3,
# train_loss,train_loss_q0,train_loss_q1,train_loss_q2,train_loss_q3
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/tmp/openai-2025-02-13-01-08-12-739876/progress_140k.csv')

# Plot train_loss and train_acc@1 over steps
plt.figure(figsize=(12, 6))

# Plot Loss
plt.subplot(2, 1, 1)
plt.plot(df['step'], df['train_loss'], label='Training Loss', color='red')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training Loss Over Steps')
plt.legend()

# Plot Accuracy
plt.subplot(2, 1, 2)
plt.plot(df['step'], df['train_acc@1'], label='Training Accuracy @1', color='blue')
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Training Accuracy @1 Over Steps')
plt.legend()

# Plot Accuracy
plt.subplot(2, 1, 2)
plt.plot(df['step'], df['train_acc@1'], label='Training Accuracy @1', color='blue')
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Training Accuracy @1 Over Steps')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot all metrics separately
for column in df.columns:
    if column not in ['step', 'samples']:  # Exclude non-metric columns
        plt.figure(figsize=(8, 4))
        plt.plot(df['step'], df[column], label=column)
        plt.xlabel('Step')
        plt.ylabel(column)
        plt.title(f'{column} Over Steps')
        plt.legend()
        plt.show()

# Plot all loss metrics together
loss_columns = [col for col in df.columns if 'loss' in col]
plt.figure(figsize=(10, 6))
for col in loss_columns:
    plt.plot(df['step'], df[col], label=col)
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('All Loss Metrics Over Steps')
plt.legend()
plt.show()

# Plot all accuracy metrics together
acc_columns = [col for col in df.columns if 'acc' in col]
plt.figure(figsize=(10, 6))
for col in acc_columns:
    plt.plot(df['step'], df[col], label=col)
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('All Accuracy Metrics Over Steps')
plt.legend()
plt.show()
```
